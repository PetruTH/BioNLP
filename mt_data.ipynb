```python
import openai
from openai import AzureOpenAI  
import time
from tqdm import tqdm

endpoint =  "SECRET_ENDPOINT"
deployment = "gpt-4.1"  
subscription_key = "SECRET_KEY"

client = AzureOpenAI(  
    azure_endpoint=endpoint,  
    api_key=subscription_key,  
    api_version="2025-01-01-preview",  
)

def generate_pipeline(sentence):
    start_time = time.time()
    chat_prompt = [
        {
            "role": "system",
            "content": [
                {
                    "type": "text",
                    "text": f"Tradu textul utilizatorului: {sentence} din limba engleza în limba română. Returneaza doar textul tradus, fără alte comentarii sau explicații."
                }
            ]
        }
    ] 
        
    messages = chat_prompt  

    completion = client.chat.completions.create(  
        model=deployment,  
        messages=messages,  
       
    )

    return completion.choices[0].message.content


```

```python
import os

depression_texts = []

path = "reddit_depression"
files = os.listdir(path)

for file in files:
    with open(os.path.join(path, file), 'r', encoding='utf-8') as f:
        text = f.read()
        all_sentences = text.strip().split('\n')
        depression_texts.extend(all_sentences)
```

```python
output_file = "depresieGOLD\\translated_DEPRESSION_sentences.txt"

for i, depression_text in enumerate(depression_texts, 1):
    if i % 100 == 0:
        print(f"Processing sentence {i} out of {len(depression_texts)}...")
    if depression_text.strip():  
        translated_text = generate_pipeline(depression_text)
        with open(output_file, "a", encoding="utf-8") as f:
            f.write(f"{i}. {translated_text}\n")
    else:
        pass
```
